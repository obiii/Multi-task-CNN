```python
import pickle
import pandas as pd
import cv2
import os
from sklearn.preprocessing import LabelEncoder
import imutils
import numpy as np
import time
from keras.models import load_model
import glob
```

```python
videoAddr = '../New Data 2020/02Edited/7248.mp4'
tempFolder = '../temp'
modelPath = '../Models/MTL/'
h5File = 'Model_loss_ES_40Ep.h5.'
jsonFile = 'Model_loss_ES_40Ep.json'
```

```python
def load_model(path, h5Name,jsonName):
    from keras.models import model_from_json
    with open(path+jsonName, 'r') as f:
        model = model_from_json(f.read())
    model.load_weights(path+h5Name)
    
    return model
```

```python
def classify(model,image):
    # load the input image and then clone it so we can draw on it later
    # our model was trained on RGB ordered images but OpenCV represents
    # images in BGR order, so swap the channels, and then resize to
    # 224x224 (the input dimensions for VGG16)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))
    # convert the image to a floating point data type and perform mean
    # subtraction
    image = image.astype("float32")/ 255.
    #mean = np.array([123.68, 116.779, 103.939][::-1], dtype="float32")
    #image -= mean
    
    # pass the image through the network to obtain our predictions
    pfr,ft = model.predict(np.expand_dims(image, axis=0))
    
    return pfr,ft
```

```python
def labelDecoder(label,cls):
    encoder = LabelEncoder()
    if cls == 'PFR':
        encoder.classes_ = np.load('../Models/Labeller/pfrClasses.npy',allow_pickle=True)
    elif cls == 'FT':
        encoder.classes_ = np.load('../Models/Labeller/ftClasses.npy',allow_pickle=True)
    elif cls == 'NOX':
        encoder.classes_ = np.load('../Models/Labeller/noxClasses.npy',allow_pickle=True)
    trueLab = encoder.inverse_transform(label)
    return(trueLab)
```
